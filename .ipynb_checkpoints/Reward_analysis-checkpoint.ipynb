```python
def plot3d_reward(x, y, r):
    X, Y = np.meshgrid(x, y)
    Z = r(X, Y)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X,Y,Z)
    return ax
```

```python
def plot_track(t):
    ratio = (lambda x,y: x / y)(*t.size())
    fig = plt.figure(figsize=(16,16/ratio))
    ax = fig.add_subplot()
    ax.plot(*t.center_line.T, 'b')
    ax.plot(*t.outer_border.T, 'b')
    ax.plot(*t.inner_border.T, 'b')
    return ax
```

```python
key = lambda wpt: [int(round(c*2))/2.0 for c in wpt]
keys = np.array([key(w) for w in track.center_line])
```

```python
ax = plot_track(track)
ax.plot(*keys.T, 'gx-')
```

```python
from rewards.hamilton import reward_function
import math
```

```python
def df2params(frame, track: Track, mapping=None):
    if mapping is None:
        mapping = {'on_track':'all_wheels_on_track', 'throttle':'speed', 'yaw':'heading', 'steer':'steering', 'track_len':'track_length'}
    d = {'closest_waypoints': [frame['closest_waypoint'], (frame['closest_waypoint']+1)%len(track.center_line)]}
    d.update({'is_reversed':False, 'waypoints':track.center_line})
    d.update({mapping[k]:frame[k] for k in mapping})
    return d
```

```python
df['new_reward'] = [math.log(1 + reward_function(df2params(r, track))) for _,r in df.iterrows()]
```

```python
df['log_reward'] = [math.log(1 + r['reward']) for _,r in df.iterrows()]
```

```python
pu.plot_track(df, track, value_field='new_reward')
```

```python
pu.plot_track(df, track, value_field='log_reward')
```
